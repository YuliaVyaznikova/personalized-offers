```python
nan_counts = df.isnull().sum()
nan_counts_proc = nan_counts/len(df) * 100
print(nan_counts_proc[nan_counts_proc > 90])
bad_cols=nan_counts_proc[nan_counts_proc > 90].index.tolist()
```

```python
print( f" количество продуктов: {df.product_id.nunique()}")
print(df.product_id.value_counts())
```

```python
def from_to_date(indx, df):
    print(f"id = {indx} дана начала: {df.timestamp[df.product_id == indx].min()} дата конца : {df.timestamp[df.product_id == indx].max()}")
```

```python
for i in range(19):
    from_to_date(i, df)
```

Посмотрим, сколько принятий у каждого продукта и его процент

```python
def prod_check(df, indx):
    print(f"indx = {indx}")
    print(f"    продано {len(df[(df.product_id == indx) & (df.is_sold == 1)])*100//len(df[df.product_id == indx])}%, не продано {len(df[(df.product_id == indx) & (df.is_sold == 0)])*100//len(df[df.product_id == indx])}%")
    print(f"    продано {len(df[(df.product_id == indx) & (df.is_sold == 1)])} шт, не продано {len(df[(df.product_id == indx) & (df.is_sold == 0)])} шт")
```

```python
for i in range(19):
    prod_check(df, i)
```
